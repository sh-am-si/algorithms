```python
import numpy as np

num = 27

arr = np.random.randint(10, 200, num)
arr = np.sort(arr)
arr
```

```python
def binary_search(arr: np.ndarray, start: int, end: int, value: int) -> int:

    while start <= end:
        mid = (start + end) // 2
        mid_value = arr[mid]
        if mid_value == value:
            return mid
        elif mid_value < value:
            start = mid + 1
        else:
            end = mid - 1

    return -1
```

```python
binary_search(arr=arr, start=0, end=arr.shape[0] - 1, value=58)
```
